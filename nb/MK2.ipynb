```python
from sympy import *
init_session()
```

```python
var('a1:17 A M E')
B = S.One
f1 = lambda p: p**a1 / (a2**a1 + p**a1)
f2 = lambda p: a4 * a5**a3 / (a5**a3 + p**a3)
p = lambda E, a15: (a14 * E * B) / a15
dA = f1(p(E,a15))*(a6+a7*M)-a8*A-a9*A**2
dM = a10*f2(p(E,a15))*A-f1(p(E,a15))*a16*a7*M-a11*M
dE = a12*(1-f2(p(E,a15)))*A-a13*E
```

```python
X = Matrix([dA, dM, dE])
Y = Matrix([A, M, E])
J = X.jacobian(Y)
```

```python
substitutions = {
    a1: Integer(2),
    a2: Integer(2),
    a3: Integer(3),
    a4: Rational(7, 10),
    a5: Integer(1),
    a6: Rational(2, 100),
    a7: Integer(20),
    a8: Integer(1),
    a9: Integer(1),
    a10: Integer(1),
    a11: Rational(1, 100),
    a12: Rational(1, 10),
    a13: Rational(3, 10),
    a14: Integer(50),
    a16: Rational(1, 10)
}
```

```python
Jprime = J.subs(substitutions)
```
